A simple program to plot chi-squared distributions. It accepts user input for number of degrees of freedom.


```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
# setting the values of degrees of freedom
while True:
    try:
        # get input
        n = int(input('Enter degrees of freedom: '))  # degrees of freedom

        if n >= 1:
            break
        else:
            print('Enter positive integer for degrees of freedom!')

    except Exception as e:
        print(e)
        print('Enter positive integer for degrees of freedom!')

print('\nChi-squared distribution - degrees: %d' % (n))
```

```python
# randomly sample n sets of points from a standard normal distribution
# and square them and add them across the degrees of freedom
samples = 10000000
sq_sum = ((np.random.normal(0, 1, [samples, n])) ** 2).sum(axis=1)

# plotting the graph
# break the distribution into counts and bins
counts, bins = np.histogram(sq_sum, bins=100)
# plot a line - opting for this instead of a histogram
plt.plot(bins[1:], counts/samples)
plt.ylabel('Probability')
plt.xlabel('x')
plt.title('Chi-squared distribution - degrees: %d' % (n))
# setting the x axis limit at 75 percentile to avoid a long tail
plt.xlim([0, np.percentile(bins, 75)])
plt.tight_layout()
plt.grid()
plt.show()
```
