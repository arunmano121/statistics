A program to study Cauchy distributions for a variety of input parameters.

```python
# import necessary libraries
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
# setting the values of location and scale parameters
while True:
    try:
        # get input
        loc = float(input('Enter location parameter: '))
        break
    except Exception as e:
        print(e)

while True:
    try:
        # get input
        scale = float(input('Enter scale parameter: '))
        break
    except Exception as e:
        print(e)

print('\nCauchy distribution parameters:')
print('Location: %0.2f Scale: %0.2f' % (loc, scale))
```

```python
# setting x-range with 10 units around the location parameter
x_val = np.linspace(loc-5, loc+5, 200)

# Cauchy function
y_val = 1/math.pi * scale / (scale**2 + (x_val - loc)**2)
```

```python
# plot the distribution
plt.plot(x_val, y_val, label='PDF')
plt.legend()
plt.ylabel('Probability')
plt.xlabel('X')
plt.xlim([loc-5, loc+5])
plt.title('Cauchy distribution - Location: %0.2f Scale: %0.2f'
          % (loc, scale))
plt.grid()
plt.tight_layout()
plt.show()
```
