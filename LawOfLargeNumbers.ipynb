A simple program to study the law of large numbers. A fair 6-sided dice roll is simulated for a variety of draws between 1 to 10000. The expected value for this event is 1/6*(1+2+3+4+5+6) = 3.5. As the number of draws increases, it is shown that the mean value of the draws converges to this value.

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# initialize empty array to hold mean value
mean_val = np.empty([0], dtype=float)

# different trials between 1 and 100000 in steps of 10
for i in np.arange(1, 100000, 10):
    # draw i numbers between 1 and 6
    this_mean = np.mean(np.random.randint(1, 7, size=i))
    mean_val = np.append(mean_val, this_mean)
    # print(i, this_mean)
```

```python
# plot convergence of the mean value as a function of draws
plt.semilogx(np.arange(1, 100000, 10), mean_val)
plt.title('Mean value vs. Number of draws', fontsize=12)
plt.xlabel('No. of Draws', fontsize=12)
plt.ylabel('Mean value', fontsize=12)
# plt.ylim([1, 6])
# the expected value for dice roll with 6 sides is 3.5
plt.axhline(3.5, color='r', linestyle='dashed')
plt.legend(['Mean Value', 'Expected Value'])
plt.tight_layout()
plt.grid(True)
plt.show()
```
