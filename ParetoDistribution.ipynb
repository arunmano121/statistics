A program to study Pareto distributions for a variety of input parameters.

```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
# setting the values of alpha
while True:
    try:
        # get input
        alpha = float(input('Enter alpha for Pareto distribution: '))
        if alpha >= 0.0:  # make sure alpha is positive
            break
    except Exception as e:
        print(e)
        print('Enter positive float for alpha!')

print('\nPareto distribution - alpha: %0.2f' % (alpha))
```

```python
# setting x-range
x_val = np.linspace(1e-3, 5, 100)

# Pareto function
xm = 1
y_val = (alpha * (xm ** alpha)) / (x_val ** (alpha+1))
y_val[x_val <= xm] = 0  # as per Pareto distribution

# cumulative distribution
y_cum = [np.sum(y_val[0:i+1]) for i in range(len(y_val))]
# divide by sum to calculate percentiles
y_cum = y_cum / y_cum[-1]
```

```python
# plot the distribution
plt.plot(x_val, y_val, label='PDF')
plt.plot(x_val, y_cum, label='CDF')
plt.legend()
plt.ylabel('Probability')
plt.xlabel('X')
plt.xlim([0, 5])
plt.title('Pareto distribution - alpha: %0.2f, $x_m$: %0.2f' % (alpha, xm))
plt.grid()
plt.tight_layout()
plt.show()
```
