A program to study Cauchy and Normal distributions for a variety of input parameters. This program also shows that Cauchy distributions have heavier tails than Normal distributions. One assumption this program makes while it is not strictly true is that the location and scale parameter for the Cauchy distribution is treated on the same terms as the mean and standard deviation of the Normal distribution.

```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
# setting the values of location and scale parameters
while True:
    try:
        # get input
        loc = float(input('Enter location parameter: '))
        break
    except Exception as e:
        print(e)

while True:
    try:
        # get input
        scale = float(input('Enter scale parameter: '))
        break
    except Exception as e:
        print(e)

print('\nCauchy and Normal distribution parameters:')
print('Location: %0.2f Scale: %0.2f' % (loc, scale))
```

```python
# setting x-range with 20 units around the location parameter
x_val = np.linspace(loc-10, loc+10, 200)

# Cauchy function
y_val_cauchy = 1/math.pi * scale / (scale**2 + (x_val - loc)**2)

# Normal function
num = (x_val - loc)/scale
y_val_normal = 1/(scale * math.sqrt(2*math.pi)) * np.exp(-0.5*(num**2))
```

```python
# plot the distribution
plt.plot(x_val, y_val_cauchy, label='Cauchy')
plt.plot(x_val, y_val_normal, label='Normal')
plt.legend()
plt.ylabel('Probability')
plt.xlabel('X')
plt.xlim([loc-10, loc+10])
plt.title('Cauchy and Normal distribution - Location: %0.2f Scale: %0.2f'
          % (loc, scale))
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# study the tail effects for x values that are between the location and
# 10 standard deviations away
x_range = np.linspace(loc, loc+10*scale, 50)
y_cauchy = 1/math.pi * scale / (scale**2 + (x_range - loc)**2)
num = (x_range - loc)/scale
y_normal = 1/(scale * math.sqrt(2*math.pi)) * np.exp(-0.5*(num**2))

# Ratio between Cauchy and Normal
ratio = y_cauchy / y_normal

# Based on the ratios, it is evident that Cauchy distributions have
# heavier tails than Normal distributions
print('x_val \t y_cauchy \t y_normal \t ratio')
for i in range(len(x_range)):
    print('%0.2f \t %0.2e \t %0.2e \t %0.2e' %
          (x_range[i], y_cauchy[i], y_normal[i], ratio[i]))
```

It is evident that ratio of Cauchy and Normal distribution at around 10 standard deviations from the normal are about $10^{19}$ indicating that the Cauchy distributions have very heavy tails. 

```python
# zoom into the distribution around the 3-4 standard deviation zone to emphasize the heavier tails
# the further we move away from the mean (location), the more flatter Cauchy distribution becomes while 
# the normal distribution falls faster
# plot the distribution 

# setting x-range around the tail end
x_val = np.linspace(loc+3*scale, loc+4*scale, 200)

# Cauchy function
y_val_cauchy = 1/math.pi * scale / (scale**2 + (x_val - loc)**2)

# Normal function
num = (x_val - loc)/scale
y_val_normal = 1/(scale * math.sqrt(2*math.pi)) * np.exp(-0.5*(num**2))

plt.plot(x_val, y_val_cauchy, label='Cauchy')
plt.plot(x_val, y_val_normal, label='Normal')
plt.legend()
plt.ylabel('Probability')
plt.xlabel('X')
plt.title('Cauchy and Normal distribution - Location: %0.2f Scale: %0.2f'
          % (loc, scale) + '\n between 3-4 SD from mean (location)')
plt.grid()
plt.tight_layout()
plt.show()
```
